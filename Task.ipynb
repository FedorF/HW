```python
import numpy as np
x = np.arange(-10, 10, 0.04)
y = np.array(2-0.5*x + np.random.randn(len(x))/10)
print x
print y
```

```python
from scipy import optimize
def f(k):
    sum = 0;
    end = len(x)
    for i in range(0, end - 1):
        sum += (y[i]-k[0]*x[i]-k[1])**2
    return sum

print f([-0.5, 2])
```

```python
result = optimize.minimize(f, [0, 0])
print(result.x)
```

```python
from matplotlib import pylab as plt
plt.plot(x, y)
y_predict = x*result.x[0] + result.x[1]
plt.plot(x, y_predict, 'b')
y_real = -0.5*x + 2
plt.plot(x, y_real, 'r')
plt.show()
```
